````
# Cole aqui o código que resolve a questão 4

# Incializando lista para armazenar os números digitados e variável para armazenar o
# atual número digitado
lista = []
numero = 0

# Obtendo números do usuário e os adicionando na lista até que -1 seja informado
while (numero >= 0):
    numero = int(input("Digite um número: "))
    if (numero >= 0):
        lista.append(numero)

# Incializando uma variável para armazenar a média final e outra para armazenar quantos
# números compõe a média
media = 0
numeros_media = 0

# Navegando em cada elemento da lista
for item in range(len(lista)):
    # Se o elemento atual for um número maior ou igual a 25 e menor ou igual a 121, ele é
    # somado à variável de média e é somado 1 à variável que contabiliza a quantidade de
    # números que compõe a média
    if (lista[item] >= 25 and lista[item] <= 121):
        media += lista[item]
        numeros_media += 1

# Se a quantidade de números que compõe a média for maior que zero, a média é calculada e
# apresentada. Caso contrário, apresenta-se que não há como calcular a média
if (numeros_media > 0):
    media /= numeros_media
    print(f"Média: {media:.2f}")
else:
    print("Não há números entre 25 e 121 para calcular a média.")
````

6.   Faça um programa que lê o gabarito de uma prova objetiva de dez questões (a resposta de cada questão é "a", "b", "c", "d" ou "e") e repetidamente lê o nome de um aluno e suas respostas para cada uma das questões, até que um nome vazio seja informado. Cada acerto vale um ponto. Ao final, indique quais alunos passaram na prova (nota >= 6), em ordem decrescente de notas.

10 - OK

````
# Cole aqui o código que resolve a questão 6

# Inicializando lista para armazenar gabarito da prova e variável para armazenar a
# questão atual
gabarito = []
questao = 1

# Obtendo gabarito de cada uma das 10 questões
while (questao <= 10):
    resposta_correta = input(f"Digite o gabarito da questão {questao}: ").lower()
    # Se o gabarito da questão informado for uma letra entre 'a' e 'e', armazena-se o mesmo
    # na lista "gabarito" e soma-se 1 para ir para a questão seguinte. Caso contrário,
    # apresenta-se um erro
    if (resposta_correta >= 'a' and resposta_correta <= 'e'):
        gabarito.append(resposta_correta)
        questao += 1
    else:
        print("Digite uma letra válida, entre 'a' e 'e'!\n")

# Inicializando lista para armazenar os alunos e suas notas e uma variável para obter o
# nome do atual aluno
alunos = []
nome = "nome"

# Obtendo alunos, suas respostas e calculando a nota de cada um até que um nome vazio seja
# informado
while (nome != ''):
    # Obtendo nome do aluno
    nome = input("Digite o nome do aluno: ")
    # Inicializando variáveis para armazenar a nota e a questão atual
    nota = 0
    questao = 1

    # Obtendo resposta do aluno em de cada uma das 10 questões
    while (nome != '' and questao <= 10):
        resposta = input(f"Digite a resposta de {nome} na {questao}: ").lower()
        # Se a resposta da questão for igual à resposta armazenada no gabarito, soma-se
        # 1 na nota e soma-se 1 para ir para a questão seguinte. Senão, se for uma letra
        # entre 'a' e 'e', armazene o mesmo na lista "gabarito" e soma-se 1 para ir para
        # a questão seguinte. Caso contrário, apresenta-se um erro
        if (resposta == gabarito[questao - 1]):
            nota += 1
            questao += 1
        elif (resposta >= 'a' and resposta <= 'e'):
            questao += 1
        else:
            print("Digite uma letra válida, entre 'a' e 'e'!\n")

    # Se as respostas das 10 questões foram informadas, adiciona-se o aluno e sua nota na
    # lista "alunos"
    if (questao > 10):
        alunos.append([nome, nota])

# Navegando na quantidade de elementos da lista "alunos" para a ordenar em ordem
# decrescente
for iteracao in range(len(alunos)):
    # Navegando em cada elemento da lista "alunos", do segundo até o último menos o número
    # de elementos já navegados pelo laço acima
    for item in range(1, len(alunos) - iteracao):
        # Se a nota do aluno atual for maior que a nota do anterior, inverte-se o conteúdo
        # de ambas as posições
        if (alunos[item][1] > alunos[item - 1][1]):
            alunos[item], alunos[item - 1] = alunos[item - 1], alunos[item]

# Apresentando alunos que passaram na prova por obter nota maior ou igual a 6
print("* Alunos que passaram na prova:")
for aluno in range(len(alunos)):
    if (alunos[aluno][1] >= 6):
        print(f"- {alunos[aluno][0]} : {alunos[aluno][1]:02d}")
````

(peso 3); "Já teve relação amorosa com a vítima?" (peso 3); "Tinha dívidas com a vítima?" (4); "Já trabalhou com a vítima?" (1). O programa deve no final emitir uma classificação sobre a participação da pessoa no crime. Se a pessoa tiver um score de 8 ou mais deve ser classificada como "Assassino"; entre 5 e 7 como "Cúmplice", entre 3 e 4 como "Suspeito". Caso contrário, ele será classificado como "Inocente".

10 - OK

````
# Cole aqui o código que resolve a questão 7

# Inicializando lista para armazenar perguntas e lista para armazenar os pesos de cada
# pergunta
perguntas = ["Telefonou para a vítima? (s/n) ",
"Esteve no local do crime? (s/n) ",
"Já teve relação amorosa com a vítima? (s/n) ",
"Tinha dívidas com a vítima? (s/n) ",
"Já trabalhou com a vítima? (s/n) "]
pesos = [2, 3, 3, 4, 1]

# Obtendo nome do usuário e inicializando variável para armazenar score obtida e variável
# para contabilizar a pergunta atual
nome = input("Digite seu nome: ")
score = 0
pergunta = 0

# Realizando perguntas e contabilizando pontuação
while (pergunta < 5):
    resposta = input(perguntas[pergunta]).lower()

    # Se resposta for  's', soma-se o peso referente à atual pergunta ao score e passa-se
    # para a próxima pergunta. Senão, se a resposta for 'n', passa-se para a próxima
    # pergunta. Caso contrário, informa-se que a opção digitada é inválida
    if (resposta == 's'):
        score += pesos[pergunta]
        pergunta += 1
    elif (resposta == 'n'):
        pergunta += 1
    else:
        print("Opção inválida!\n")

# Se o score for maior ou igual a 8, informa-se que a pessoa é o assassino. Senão, se o
# score for entre 7 e 5, informa-se que a pessoa é cúmplice. Senão, se o score for entre
# 4 e 3, informa-se que a pessoa é suspeita. Caso contrário, informa-se que a pessoa é
# inocente
if (score >= 8):
    print("Assassino")
elif (score >= 5):
    print("Cúmplice")
elif (score >= 3):
    print("Suspeito")
else:
    print("Inocente")
````

````
# Cole aqui o código que resolve a questão 8

# Instalando traceroute
!apt install -y traceroute

# Importando biblioteca subprocess para execução do comando traceroute
import subprocess

# Obtendo nome de uma máquina e armazenando comando do traceroute, inicializando variável
# para armazenar a saída da execução do comando e dividindo a string em linhas por meio da
# quebra de linha
CMD = "traceroute -n " + input("\nDigite o nome de uma máquina: ")
caminho = str(subprocess.run(CMD, capture_output=True, shell=True).stdout).split("\\n")

# Removendo primeira e última linha da lista "caminho", armazenando total de elementos
# da lista "caminho" e inicializando variável para armazenar a linha atualmente testada
caminho.pop(0)
caminho.pop(-1)
quantidade_linhas = len(caminho)
linha = 0

# Lendo conteúdo da linha, realizando divisão, análise e separação de elementos (parsing)
# e apresentando endereço IP e menor tempo para alcançá-lo
while (linha < quantidade_linhas):
    # Se for a linha da lista "caminho" for a primeira, apresenta-se legenda para os dados
    if (linha == 0):
        print("| Endereço        | Menor tempo |")

    # Separando linha da lista "caminho" por espaços, removendo o primeiro elemento da
    # sublista formada no elemento atual da lista "caminho", armazenando quantos
    # elementos há nessa lista, inicializando lista para armazenar os tempos presentes
    # na atual sublista da lista "caminho" e inicializando variável para armazezar a
    # posição do elemento da sublista atualmente testado
    caminho[linha] = caminho[linha].split()
    caminho[linha].pop(0)
    quantidade_itens = len(caminho[linha])
    tempos = []
    item = 0

    # Navegando em cada item da sublista da lista "caminho" para remover possíveis
    # asteriscos na sublista e contabilizar os tempos
    while (item < quantidade_itens):
        # Se o elemento atual da sublista for um asterisco, remove-se o mesmo e subtrai-se
        # 1 do total de elementos da sublista e da posição do elemento atualmente testado
        if (caminho[linha][item] == '*'):
            caminho[linha].pop(item)
            item -= 1
            quantidade_itens -= 1
        # Senão, se o elemento atual da sublista for "ms", converte-se o elemento anterior
        # em número e o adiciona na lista "tempos", então remove-se o "ms" e subtrai-se 1
        # do total de elementos da sublista e da posição do elemento atualmente testado
        elif (caminho[linha][item] == "ms"):
            caminho[linha][item - 1] = float(caminho[linha][item - 1])
            tempos.append(caminho[linha][item - 1])
            caminho[linha].pop(item)
            item -= 1
            quantidade_itens -= 1

        # Somando 1 para ir para a posição do próximo elemento da sublista da lista
        # "caminho"
        item += 1

    # Se a quantidade de elementos na sublista da lista "caminho" for 0, remove-se essa
    # sublista e subtrai-se 1 do total de linhas da lista "caminho". Caso contrário,
    # apresenta-se o IP e o menor dos tempos para alcançar esse endereço, armazenados na
    # lista "tempos"
    if (quantidade_itens == 0):
        caminho.pop(linha)
        quantidade_linhas -= 1
    else:
        print(f"| {caminho[linha][0]:<15} | {min(tempos):<8} ms |")

    # Somando 1 para ir para a posição da próxima linha da lista "caminho"
    linha += 1
````

9.   Faça um programa que leia uma lista de números e indique a maior soma de cinco elementos consecutivos.

````
# Cole aqui o código que resolve a questão 9

# Inicializando lista para armazenar os números digitados pelo usuário e variável para
# armazenar o número atual digitado pelo usuário
lista = []
numero = 0

# Obtendo números do usuário e os adicionando na lista até que -1 seja informado ou a lista
# possua menos de 5 elementos
while (numero >= 0 or len(lista) < 5):
    numero = int(input("Digite um número: "))
    if (numero >= 0):
        lista.append(numero)

# Inicializando variável para armazenar a maior soma de 5 números
maior_soma = 0

# Navegando nos itens da lista "lista" a partir do quinto
for item in range(4, len(lista)):
    # Armazenando soma do elemento atual da lista com os 4 anteriores
    soma = lista[item] + lista[item - 1] + lista[item - 2] + lista[item - 3] + lista[item - 4]

    # Se a soma obtida for maior que a maior soma armazenada, substutui-se a mesma pela
    # nova maior
    if (soma > maior_soma):
        maior_soma = soma

# Apresentando maior soma
print(f"A maior soma entre 5 elementos consecutivos é {maior_soma}.")
````
